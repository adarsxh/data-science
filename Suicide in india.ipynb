```python
cause = df[df.Type_code == 'Causes']
cause
```

```python
cause.Type.unique()
```

```python
plt.figure(figsize = (6,8))
sns.barplot(y = 'Type', x = 'Total', data = cause, hue = 'Gender')
#family problem is the cause that lead to more number of suicides, also there are causes where only female are doing suicide such as dowry and illegitimate pregnancy.
```

```python
upgen = up.groupby('Gender')
upgen.head()
```

```python
upgen.describe()
```

```python
upgen['Total'].agg(np.sum)
#this shows that total number of suicide is more in men than female but the at highest number of suicide at a particular year is more in female
```

```python
upyr = up.groupby('Year')
upyr.head()
```

```python
upyr.Type_code.unique()
```

```python
print(upyr['Total'].agg(np.sum).idxmax())
print(upyr['Total'].agg(np.sum).max())
#maximum suicide happened in 2011 
```

```python
print(upyr['Total'].agg(np.sum).idxmin())
print(upyr['Total'].agg(np.sum).min())
#minimum suicide happned in 2006
```
